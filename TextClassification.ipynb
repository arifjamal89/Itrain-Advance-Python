Step 1 : Load Data
---



```python
import nltk
from nltk.corpus import names
nltk.download('names')
labeled_names = ([(name, 'male') for name in names.words('male.txt')]+ [(name, 'female') for name in
names.words('female.txt')])
```

```python

import random 
random.shuffle(labeled_names) #shuffle data set
labeled_names
```

Step 2 : Feature Extractor

```python
def feature_extractor(name): return {'last_letter': name[-1]}

featureset = [(feature_extractor(name), gender) for (name, gender) in labeled_names]
featureset
```

Split Test/ Training Data

```python
train_set = featureset[500:] # 500 set data
test_set = featureset[:500]  # 500 set data
```

Step 4/5 : Load & Train Classifier

```python
import nltk

classifier =nltk.NaiveBayesClassifier.train(train_set)
```

Step 6 : Evaluation/ Prediction

```python
x=classifier.classify(feature_extractor('Ain')) #give new name as neo
print(x)
nltk.classify.accuracy(classifier, test_set)
```
