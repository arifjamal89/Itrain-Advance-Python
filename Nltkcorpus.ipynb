```python
from nltk.corpus import PlaintextCorpusReader 
corpus_root = '/content/sample_data' # location to store library
my_corpus = PlaintextCorpusReader(corpus_root,"sample.txt")

```

```python
text = my_corpus.raw(fileids="sample.txt")
```

```python
my_corpus.raw(fileids="sample.txt")
```

```python
words = text.split()
words
```

```python
distr = nltk.FreqDist(words)
distr
```

```python
distr = nltk.FreqDist(words)
distr.most_common(10)
```
